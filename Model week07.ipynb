```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Parameter distribusi binomial
n = 10  # Jumlah percobaan
p = 0.5  # Probabilitas sukses

# Nilai x (jumlah sukses)
x = np.arange(0, n+1)

# Distribusi binomial
binom_prob = binom.pmf(x, n, p)

# Plot probabilitas
plt.bar(x, binom_prob)
plt.xlabel('Jumlah Sukses')
plt.ylabel('Probabilitas')
plt.title('Distribusi Binomial: n=10, p=0.5')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameter distribusi normal
mean = 70  # Rata-rata
std_dev = 10  # Standar deviasi

# Rentang nilai x
x = np.linspace(mean - 4*std_dev, mean + 4*std_dev, 1000)

# Distribusi normal
normal_prob = norm.pdf(x, mean, std_dev)

# Plot distribusi normal
plt.plot(x, normal_prob)
plt.xlabel('Berat Badan (kg)')
plt.ylabel('Probabilitas')
plt.title('Distribusi Normal: mean=70kg, std_dev=10kg')
plt.show()
```
