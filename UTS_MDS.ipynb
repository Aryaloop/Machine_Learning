```python
import random
import matplotlib.pyplot as plt

# Ukuran persegi panjang
width = 10
height = 10

# Ukuran segitiga
base_triangle = 8
height_triangle = 8

# Fungsi untuk mengecek apakah titik berada di dalam segitiga
def is_point_in_triangle(x, y):
    # Koordinat segitiga berada di sudut kiri bawah persegi panjang
    # Persamaan batas segitiga: y <= -(height_triangle / base_triangle) * x + height_triangle
    return y <= -(height_triangle / base_triangle) * x + height_triangle

# Jumlah titik simulasi
num_points = 10000
points_in_triangle = 0

# Menyimpan koordinat titik untuk visualisasi
inside_x = []
inside_y = []
outside_x = []
outside_y = []

# Menebar titik acak dan menghitung berapa yang jatuh di dalam segitiga
for _ in range(num_points):
    x = random.uniform(0, width)
    y = random.uniform(0, height)
    if is_point_in_triangle(x, y):
        points_in_triangle += 1
        inside_x.append(x)
        inside_y.append(y)
    else:
        outside_x.append(x)
        outside_y.append(y)

# Menghitung estimasi luas segitiga
area_rectangle = width * height
estimated_area_triangle = (points_in_triangle / num_points) * area_rectangle
actual_area_triangle = 0.5 * base_triangle * height_triangle  # Luas sebenarnya dari segitiga

# Output hasil
print(f"Estimasi luas segitiga: {estimated_area_triangle:.2f} satuan kuadrat")
print(f"Luas segitiga sebenarnya: {actual_area_triangle} satuan kuadrat")
print(f"Error estimasi: {abs(actual_area_triangle - estimated_area_triangle):.2f} satuan kuadrat")

# Visualisasi hasil Monte Carlo
plt.figure(figsize=(8, 8))
plt.scatter(outside_x, outside_y, color='blue', s=1, label="Outside Triangle")
plt.scatter(inside_x, inside_y, color='red', s=1, label="Inside Triangle")

# Menggambar segitiga dan persegi panjang
plt.plot([0, base_triangle, 0, 0], [0, 0, height_triangle, 0], color="green", linewidth=2, label="Triangle")
plt.xlim(0, width)
plt.ylim(0, height)
plt.gca().set_aspect('equal', adjustable='box')
plt.title("Simulasi Monte Carlo untuk Estimasi Luas Segitiga")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

```python
from scipy.stats import norm

# Diberikan
mean_duration = 60  # rata-rata durasi dalam jam
std_dev_duration = 8  # deviasi standar dalam jam

# Bagian (a): Mencari waktu yang memastikan 95% peserta menyelesaikan kursus tepat waktu
confidence_level = 0.95
time_limit_95_percent = norm.ppf(confidence_level, mean_duration, std_dev_duration)
print(f"Waktu maksimum untuk memastikan 95% peserta selesai tepat waktu: {time_limit_95_percent:.2f} jam")

# Bagian (b): Probabilitas seorang peserta menyelesaikan kursus lebih dari 70 jam
time_threshold = 70
prob_more_than_70 = 1 - norm.cdf(time_threshold, mean_duration, std_dev_duration) 
print(f"Probabilitas peserta menyelesaikan kursus lebih dari 70 jam: {prob_more_than_70:.4f}")
```
